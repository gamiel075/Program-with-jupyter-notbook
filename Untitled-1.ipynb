```python
import pandas as pd
import matplotlib.pyplot as plt
from  colorama import Fore,Back,Style,Init



tabela = pd.read_excel("Vendas.xlsx") #para ler o arquivo panda,esse pd é o apelido que damos a biblioteca
display(tabela) #este deisplay é como se fosse um 'print'

faturamento_total = tabela["Valor Final"].sum()  #é assim que se forma o valor total a coluna de uma tabela, o 'sum()' é somar
print(faturamento_total)

if(faturamento_total < 1000000 and  faturamento_total > 300000):
    print(Fore.GREEN  + str(faturamento_total))


#faturameto por loja
faturamento_media_loja = tabela[['ID Loja',"Valor Final"]].groupby("ID Loja").sum() #aqui eu agrupei as informações utilizadas('groupby' e o 'sum' junta os respectivos nomes e seus valores)
display(faturamento_media_loja)

# Criar o gráfico de barras
plt.bar(faturamento_media_loja.index, faturamento_media_loja['Valor Final'])

# Adicionar rótulos e título
plt.xlabel('ID da Loja')
plt.ylabel('Faturamento Total')
plt.title('Faturamento Total por Loja')

# Exibir o gráfico
plt.show()


#faturamento da loja  por produto


import matplotlib.pyplot as plt

# Se não tiver o matplotlib instalado, você pode instalá-lo usando: pip install matplotlib

# Seu código para calcular o faturamento por produto por loja
faturamento_por_produto = tabela[['ID Loja', 'Produto', 'Valor Final']].groupby(['ID Loja', 'Produto']).sum()

# Escolher uma loja específica para o exemplo (você pode adaptar conforme necessário)
loja_especifica = 'Iguatemi Campinas'
dados_loja_especifica = faturamento_por_produto.loc[loja_especifica]

# Criar o gráfico de pizza
plt.pie(dados_loja_especifica['Valor Final'], labels=dados_loja_especifica.index, autopct='%1.1f%%', startangle=140)

# Adicionar título
plt.title(f'Distribuição do Faturamento por Produto para a {loja_especifica}')

# Exibir o gráfico
plt.show()







```
