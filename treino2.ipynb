```python
import pandas as pd
import matplotlib.pyplot as plt
from colorama import Fore,Back,Style, init
from colorama import Fore,Back,Style, init


tabela = pd.read_excel('VENDASV2.xlsx')
display(tabela)

faturamento_total = tabela['Valor_Vendas'].sum()

if(faturamento_total < 1000000 and  faturamento_total > 300000):
    print(Fore.GREEN + Style.BRIGHT + str(faturamento_total))



faturamentoxloja = tabela[['Loja','Valor_Vendas']].groupby('Loja').sum()
display(faturamentoxloja)
cor = ['red','blue','green']
plt.bar(faturamentoxloja.index,faturamentoxloja['Valor_Vendas'],color = cor)
plt.xlabel('Loja')
plt.ylabel('Valor_Final')
plt.title('Faturamento por Loja')
plt.show()

faturamentoxlojaxproduto= tabela[['Categoria','Valor_Vendas']].groupby('Categoria').sum()
display(faturamentoxlojaxproduto)
cor = ['red','blue','yellow','pink']
plt.bar(faturamentoxlojaxproduto.index , faturamentoxlojaxproduto['Valor_Vendas'],color=cor)
plt.xlabel('Categoria')
plt.ylabel('Valor_Final')
plt.title('Faturamento por Categoria')
plt.show()


faturamento_por_produto = tabela[['Loja','Produto',"Valor_Vendas"]].groupby(["Loja",'Produto']).sum() 
display(faturamento_por_produto)

loja_1 = 'A'
dados_loja1 = faturamento_por_produto.loc[loja_1]
plt.pie(dados_loja1['Valor_Vendas'], labels=dados_loja1.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribução das Vendas Loja A')
plt.show()

loja_2 = 'B'
dados_loja2 = faturamento_por_produto.loc[loja_2]
plt.pie(dados_loja2['Valor_Vendas'], labels=dados_loja2.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribução das Vendas Loja B')
plt.show()

loja_3 = 'C'
dados_loja3 = faturamento_por_produto.loc[loja_3]
plt.pie(dados_loja3['Valor_Vendas'], labels=dados_loja3.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribução das Vendas Loja C')
plt.show()


#a procima etapa será inserir os indicadores de faturamento,recita e etc,meta por loja,metaxhora e etc,deficit

```
