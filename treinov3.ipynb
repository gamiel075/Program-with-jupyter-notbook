```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display






tabela = pd.read_excel('Pasta1.xlsx')
display(tabela)

df = tabela   #formula para orgazizar a tabela em forma de ranking
df_ranking = df.sort_values(by= 'Valor_Venda',ascending=False)
display(df_ranking)

plt.bar(df_ranking.index,df_ranking['Valor_Venda'])
plt.xlabel('Vendedores')
plt.ylabel('renda')
plt.title('ranking')
plt.show()

faturamento_total= tabela['Valor_Venda'].sum()
quantidade_vendas = tabela['Qtd_Vendas'].sum()
print(quantidade_vendas)
media_venda= faturamento_total / quantidade_vendas
media_venda_vendedor = faturamento_total / 15
media_vendas = quantidade_vendas / 15



tabela2 = {
    'Indicador': ['Faturamento Total', 'Média de lucro por venda', 'Média venda por Pessoa' , 'quantidade de vendas realizada','quantidade feitas por pessoa'],
    'Valor': [faturamento_total, media_venda, media_venda_vendedor,quantidade_vendas, media_vendas]
}

df= pd.DataFrame(tabela2)

df_rounded = df.round(2)
display(df_rounded)





```
