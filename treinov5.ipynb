```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display 

tabela = pd.read_excel('vendasv3.xlsx')
display(tabela)

tabela_arrecadação_mês = tabela[['Mês','Lucro']]
display(tabela_arrecadação_mês)

plt.bar(tabela_arrecadação_mês.index,tabela_arrecadação_mês['Lucro'])
plt.xlabel('Mês')
plt.ylabel('Lucro')
plt.xticks(tabela_arrecadação_mês.index, tabela_arrecadação_mês['Mês'])
plt.show()

tabela_total_produtos = tabela[['Mês','Vendas']]
display(tabela_total_produtos)


dados = tabela_total_produtos['Vendas']
plt.pie(dados, labels=tabela_total_produtos['Mês'], autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de total de vendas por m~es')
plt.show()


faturamento_total = tabela['Lucro'].sum()
total_vendas = tabela['Vendas'].sum()
media_venda_mes = total_vendas / 4
media_lucro_mes = faturamento_total /4

tabela2 = {
    'Indicador': ['faturamento total','total de vendas','media de vendas','media lucro'],
    'valor' : [faturamento_total,total_vendas,media_venda_mes,media_lucro_mes]

}

df = pd.DataFrame(tabela2)
display(df) 











```
