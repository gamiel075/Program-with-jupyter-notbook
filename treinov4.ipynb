```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

# Dados da tabela
data = {
    'Mês': ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio'],
    'Produto A (em unidades)': [150, 180, 160, 200, 170],
    'Produto B (em unidades)': [200, 220, 180, 250, 210],
    'Produto C (em unidades)': [120, 100, 130, 110, 140],
    'Receita Total ($)': [8500, 9200, 8700, 9400, 9000]
}
df = pd.DataFrame(data)
display(df)

cor = ['red','blue','yellow','green','black']
plt.bar(df.index,df['Receita Total ($)'],color=cor)
plt.xlabel('mês')
plt.ylabel('valor')
plt.title('Renda por mês')
plt.show()

#desafio é saber quanto

faturamento_total = df['Receita Total ($)'].sum()
print(faturamento_total)

qtd_vendasA = df['Produto A (em unidades)'].sum()
qtd_vendasB = df['Produto B (em unidades)'].sum()
qtd_vendasC = df['Produto C (em unidades)'].sum()
qtd_total_vendas = qtd_vendasA + qtd_vendasB + qtd_vendasC

tabela1 = { 'Produtos':['A','B','C'], 
            'Quantidade' : [qtd_vendasA,qtd_vendasB,qtd_vendasC]
           }
df2 = pd.DataFrame(tabela1)
display(df2)

média_mês_faturamento = faturamento_total / 5
media_produto_vendas = qtd_total_vendas / 5

tabela2 = {'Indicador' : ['total de vendas','faturamento total' , ' media de vendas', 'media de renda'],
           'valor':[qtd_total_vendas,faturamento_total,media_produto_vendas,média_mês_faturamento]

}

df3 = pd.DataFrame(tabela2)
display(df3)





```
