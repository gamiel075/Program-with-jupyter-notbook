````python
renda_turno = tabela[['Turno','venda)valor)']].groupby('Turno').sum()
display(renda_turno)
````

## O turno que mais vende.

Descobrimos também que o o período da manhã é o período que mais vende,podendo ser oque as pessoas mais comprarm, *mas não* ,ou melhor mais ou menos. Se a compra foi online, o horário de compra pode ser diverso, *_uma outra tabela com as compras e vendas é necessária._*

````python
renda_categoria = tabela[['Aréa','venda)valor)']].groupby('Aréa').sum()
#NO CASO SE VOCÊ QUISSECE COLOCAR EM ORDEM CRESCENTE 
df2 = renda_categoria.sort_values(by='venda)valor)', ascending=False)
display(df2)

````

Observamos agora que a área de Moda é a campeã de vendas,o carro chefe.

````python
renda_categoria_loja = tabela[['Equipe','Aréa','venda)valor)']].groupby(['Equipe','Aréa']).sum()
display(renda_categoria_loja)
````

## Novas descobertas

Aqui descobrimos quem nem todas as lojas vendem todos os produtos,alguns tipos de produtos se destancam em uma e em outras não.

````python
renda_loja = tabela[['Equipe','venda)valor)']].groupby(['Equipe']).sum()
df3 = renda_loja.sort_values(by = 'venda)valor)' , ascending= False)
display(df3)
````

Vemos aqui agora em forma de ranking as lojas que mais vendem.Quais estrátegias cada loja aplica e qual são tipos de produtos mais vendidos respectivamente.
